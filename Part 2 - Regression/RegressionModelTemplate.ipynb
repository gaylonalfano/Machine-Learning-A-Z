## Python Non-Linear Regression Template and Evaluating Regression models

### R Squared
Want to minimize the sum of squares of residuals. Want to find best fit (linear regression). As you minimize SSres it goes smaller. So 1-SSres, becomes greater/larger. That case R Squared is equal to 1. The closer to 1, the better.

### Adjusted R Squared
When you add more independent variables to your model, your simple R^2 value will **never** decrease. Therefore, you need to use Adj. R^2 for goodness of fit evaluation. It penalizes you for adding variables that don't help your model.

### Interpreting Linear Regression Coefficients
**Sign(+/-):** If the sign is positive, then if you increase this independent var, then your dependent will increase. If the sign is negative, then it's the opposite effect. Remember that coefficients portray the additional effect that every single variable brings into the model.

**Magnitude:** Tricky. Must remember that magnitude should be discussed in terms of: for every dollar increase in R&D spend, will drive a .79 cents units of profit.
