## Logistic Regression Intuition - LR is a linear classifier! LR returns probabilities.

Now what can we do with LR? We can use it to predict probabilities (p-hat p^). Going to be very important/used a lot when we build "jusizmagraphic"?? segmentation. It's better than just having a 1 or 0. It can predict a binary outcome.

It's still trying to fit the best line/curve to our data. The same as linear regression. If you want a predicted dependent variable, y^, then you typically select an arbitrary line (usually 50% - or based on whatever knowledge you may have regarding the problem at hand).

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
# Handy tip: X must be a DF/Matrix, not a Series/Vector
# Therefore use .iloc[:, 1:2] instead of [:, 1]
dataset = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:, 2:4].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
X
type(X)
X.shape
```

```python
y
type(y)
y.shape
```

```python
# Splitting the dataset ino the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
type(y_test)
type(X_test)
X_train
```

```python
# Feature Scaling - Needed in this case because we want
# accurate predictions (which users will buy the luxury SUV)
# This transforms them to be scaled. Don't have to scale 
# y_train since it's already the categorical variable.
# X_train is well trained. We fitted sc_X StandardScaler to our matrix of features
# X_train and we used the same object on X_test.
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler() 
X_train = sc_X.fit_transform(X_train) 
X_test = sc_X.transform(X_test)
```

```python
X_train
X_test
```

```python
# Fitting the Logistic Regression model to the dataset
from sklearn.linear_model import LogisticRegression
# We then create a new LOGISTIC REGRESSION object from this class is going 
# to be our classifier that we will fit on our training set. 
classifier = LogisticRegression(random_state = 0)

# Fit to training set. This means that we're taking our logistic regression object
# and fitting to the training set so that our classifier learns the 
# correlations between X_train and y_train. Then by leraning the correlations
# can use what it learn to predict new observations. We will then 
classifier.fit(X_train, y_train) 
```

```python
# Predicting the Test set results. 
# y_pred is the vector of predictions of each of the X_test set observations
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# Making the Confusion Matrix. This time we're pulling in a function
# instead of a class. 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

## Confusion Matrix
After calling the confusion_matrix function with the y_test, y_pred arguments, the results (see below) basically tell us we have 65 + 24 (89) CORRECT predictions and 3 + 8 (11) INCORRECT predictions. So, essentially 89% probability of accurately predicting whether a new user on the social network, based on their age and estimated salary independent variables, would purchase the luxury SUV.

```python
# 65 and 24 are the number of correct predictions. 
# 3 and 8 are the incorrect predictions
cm
```

## Next we evaluate the model performance with a visualization
We'll clearly see the decision regions on when/when not a purchase is predicted to be made. This map is a little complicated...

All the points we see on the graph (red/green) are the observation points of our TRAINING set. Each users are characterized by their age and estimated salary. The red points are the training set observations for which dependent variable PURCHASED is equal to zero. The green points are the training set observations where the dependent variable PURCHASED is equal to one.

We can see that the users who are young/lower salary didn't buy it and vice-versa for the older/richer. 

What is the goal of classification? We want to classify (or make a classifier that will catch) the right users into the right categories, and we represented the way our classifier catches these users by plotting the prediction regions (red and green). The prediction regions catch all the users who do/do not purchase the SUV. 

Be careful! This is according to the classifier! This is just a prediction. But we can compare the predictions to the "ground truth" by looking at the points on the graph, which represent actual values.

This type of analysis can help this company since, for each NEW USER of the social network, our classifier can tell based on their age and estimated salary if this user belongs to the red or green regions. This way the business client company can substantially optimize this marketing compaign by targeting the social network ads to the users in the GREEN region, since these are the users who are predicted to purchase the SUV. 

The "straight line" is called the prediction boundary separator, because it's the boundary between the two prediction regions. It's meant to be straight since our classifier is a linear classifier. That means that we have two dimensions (independent vars),  age and estimated salary, the separator will always be a straight line. If we had three independent variables, then it will be a straight plane (?? couldn't hear that well), separating two spaces. When we build non-linear classifiers, the prediction boundary separator will be something else (taught later). 

Need to remember that this is the training set. The classifier was trained based on the training set. Next is to see how well it does with new observations (with the test set). 

```python
# Visualizing the Training Set results
from matplotlib.colors import ListedColormap
plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = 8, 4
%matplotlib inline

X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.25, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
               c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Trainin Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Visualizing the Test Set results. Just need to replace X_train, y_train
# with X_test, y_test and that's it! 
from matplotlib.colors import ListedColormap  # in order to color the data points
plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = 8, 4
%matplotlib inline

X_set, y_set = X_test, y_test  # just convenient local vars so don't have to swap all of them when doing both train and test

# meshgrid() allows us to create the size of the graph and pixel size
# The start/stop params: take the min AGE value substract one so the pixels don't "squeeze" against axis
# Take the max AGE value plus one so not to squeeze again (it's a range basically)
# Do the same for min/max of estimated salary so not crowd the axes.
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))

# This is where the magic happens. We apply the classifier on all the pixel points
# and colors accordingly. 
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.25, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())  # plot limits of Age
plt.ylim(X2.min(), X2.max())  # plot limits of Estimated Salary

# Plot all the data points / observations / real values (the red/green dots)
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
               c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()  # To specify to display the graph and this is the end of the plot
```

## For the above plot (Test Set)
Looks good. These are actual results: 89 correct, 11 incorrect predictions based on the confusion matrix (see further above).

## So how did we plot these regions in a plot?
We took all the pixels on this entire framework (i.e., the actual background resolution of the entire graph with a 0.01 resolution -- see parameter "step = 0.01"), and we applied our classifier on it. So, like each pixel represents a made up user (almost like a new observation but ones that are not in our original dataset) with an age and estimated salary. 

For each pixel point, we applied our logistic regression model to that pixel point (so it has an age and salary value) in order to predict if that point has value 0 or 1. If it's 0, then it colors the pixel point in red and vice versa. 

It basically colors all the points that have classifier value of 0 or 1.
